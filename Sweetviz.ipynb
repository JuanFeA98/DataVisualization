## Sweetviz

### Importamos nuestras librerias

```python
import pandas as pd
import sweetviz as sv
```

### Cargamos y preparamos nuestros datos

```python
df_train = pd.read_csv('./Data/titanic_train.csv')
df_test = pd.read_csv('./Data/titanic_test.csv')
```

```python
df_train.sample(5)
```

```python
df_test.sample(5)
```

```python
my_report_comp= sv.compare([df_train, "Train"], [df_test, "Test"], "Survived")
```

```python
my_report_comp.show_html("Reporte_comp.html")
```
